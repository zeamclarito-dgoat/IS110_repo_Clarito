#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hour = float(input("enter a hour:"))
rate = float(input("enter a rate:"))

if hour > 40:
    overtime = hour - 40
    pay = (40 * rate) + (overtime*rate*1.5)
else :
    pay  = hour * rate
print("pay:",pay)
````
