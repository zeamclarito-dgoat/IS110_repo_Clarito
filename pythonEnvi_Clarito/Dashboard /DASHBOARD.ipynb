```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import plotly.express as px
print('Done')
```

```python
np.random.seed(42)
random.seed(42)
first_names = [
    'Francis' , 'Kyza' , 'Phoenix' , 'Hanni' , 'Feneo' , 'Maria'
]
last_names = [
    'Tupaz' , 'Gloria' , 'Ballego' , 'Villa' , 'Abrot' , 'Diesto'
]
middle_initial = ['B.' , 'M' , 'G' , 'F' , 'F' , 'B']
address    = ['Cabatuan', 'San Dio', 'St. Barbara', 'Mohon' , 'Maasin' , 'Iloilo']
birthday   = ['Feb 27 , 2004', 'Sep 27 , 2004', 'Jan 3                                                                  , 2001', 'Jan 4 , 2001' , 'Jan 5 , 2001' , 'Jan 6 , 2001']

school_id  = ['2023-1234-A','2023-2234-A','2023-3234-A','2023-4234-A','2023-5234-A','2023-6234-A']
```

```python
rows = []

for i in range(6):

    first_name = first_names[i % len(first_names)]
    last_name = last_names[i % len(last_names)]
    middle_initial_item = middle_initial[i % len(middle_initial)]
    full_name = f"{first_name} {middle_initial_item}. {last_name} "
    address_item = address[i % len(address)] + ", Iloilo"
    bday_item = birthday[i % len(birthday)]
    school_id_item = school_id[i % len(school_id)]
    email = f"{full_name.lower().replace(' ', '.') }@gmail.com"
    rows.append({
        'Name': full_name,
        'Address': address_item,
        'Email': email,
        'Birthday': bday_item,
        'School ID': school_id_item
    })

df = pd.DataFrame(rows)
df.to_csv('Students_Data.csv', index=False)
print(f'Generated {len(df)} rows -> Students_Data.csv')
```

```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('Students_Data.csv')

df['Birthday'] = pd.to_datetime(df['Birthday'])
df['Birth_Month'] = df['Birthday'].dt.strftime('%B')

month_counts = df['Birth_Month'].value_counts().reset_index()
month_counts.columns = ['Birth_Month', 'Count']

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

for month in month_order:
    if month not in month_counts['Birth_Month'].values:
        month_counts = pd.concat([month_counts,
                                  pd.DataFrame({'Birth_Month': [month], 'Count': [0]})],
                                 ignore_index=True)

month_counts = month_counts.set_index('Birth_Month').loc[month_order].reset_index()

fig = px.bar(
    month_counts,
    x='Birth_Month',
    y='Count',
    title='Students by Birth Month',
    labels={'Birth_Month': 'Birth Month', 'Count': 'Number of Students'},
    color='Birth_Month',
    text='Count'
)

fig.update_layout(
    xaxis_title="Birth Month",
    yaxis_title="Number of Students",
    showlegend=False,
    height=500,
    title_x=0.5,
    xaxis={'categoryorder': 'array', 'categoryarray': month_order}
)

fig.show()
```
