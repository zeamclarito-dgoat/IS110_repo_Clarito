```python
#B1
x=np.full((5,5), 1)   #it is a function used to create an array of a specified shape (as for my code it is 5x5 ) where every element is filled with the same value(as for my code it is 1)
print(x,'\n')         #prints what the array would look like
print(x.shape,'\n')   #prints the shape of the array
print(x.size,'\n')    #prints the size of the array
print(x.ndim,"\n")    #prints the number of dimension
```

```python
#C
num = np.arange(-5,6) #It is a function that evenly spread the ranges between -5 and 6 (-5<x<6)
print(num)            #prints num
b=num*2               #multiplies the number by 2
print(b)              #Prints b
```
